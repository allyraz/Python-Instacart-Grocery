### Step 4 Create loyalty flag using transfom() and loc()

#### Creating max_order column

```python
ords_prods_merge['max_order']=ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head()
```

#### Creating loyalty profile flags

```python
# Flag 1
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'loyal customer' 
```

```python
# Flag 2
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'regular customer'
```

```python
# Flag 3
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'new customer' 
```

```python
ords_prods_merge[['order_number','user_id','max_order', 'loyalty_flag']].head(30)
```

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

### Step 5 basic stats for each loyalty category

```python
ords_prods_merge.groupby(['loyalty_flag']).agg({'prices':['mean' , 'max', 'min']})
```

### Step 6 Creating spending profile flag

```python
# Creating new column for average price in orders
ords_prods_merge['ave_prices'] = ords_prods_merge.groupby('user_id')['prices'].transform(np.mean)
```

```python
# Flag 1: <10 = low spender
ords_prods_merge.loc[ords_prods_merge['ave_prices'] < 10, 'spending_flag'] = 'low spender'
```

```python
# Flag 2: >=10 = High spender
ords_prods_merge.loc[ords_prods_merge['ave_prices'] >= 10, 'spending_flag'] = 'high spender'
```

```python
ords_prods_merge[['order_number','user_id','ave_prices', 'spending_flag']].head(30)
```

```python
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

### Step 7 creating order frequency profile flag

#### Creating new column using transform & agg

```python
# create a median days column in df using transform
ords_prods_merge['median_days_since_last_order'] = ords_prods_merge.groupby('user_id')['days since_last_order'].transform(np.median)
```

```python
ords_prods_merge[['order_number','user_id','days since_last_order','median_days_since_last_order',]].head(30)
```

#### Creating frequency flags using loc

```python
# Flag 1: median > 20 = non-frequent customer
ords_prods_merge.loc[ords_prods_merge['median_days_since_last_order'] > 20, 'frequency_flag'] = 'non-frequent customer'
```

```python
# Flag 2: median <=20 and >10 = regular customer
ords_prods_merge.loc[(ords_prods_merge['median_days_since_last_order'] <= 20) & (ords_prods_merge['median_days_since_last_order'] > 10), 'frequency_flag'] = 'regular customer'
```

```python
# Flag 3: median <=10 = frequent customer
ords_prods_merge.loc[ords_prods_merge['median_days_since_last_order'] <= 10, 'frequency_flag'] = 'frequent customer'
```

```python
# cheking flag
ords_prods_merge[['order_number','user_id','days since_last_order','median_days_since_last_order','frequency_flag']].head(30)
```

```python
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```

```python
#checking NaN values
ords_prods_merge[ords_prods_merge['frequency_flag'].isna()]
```

The Nan frequency value is for customer with max_order = 1 (i.e. have only ever made 1 order)
