### Creating new function

```python
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# applying the function
df_ords_prods_merged_1['price_range'] = df_ords_prods_merged_1.apply(price_label, axis=1)
```

```python
# counting the values in the column
df_ords_prods_merged_1['price_range'].value_counts(dropna = False)
```

### Finding busiest days

```python
# finding what days are busiest
df_ords_prods_merged['order_day_of_week'].value_counts(dropna=False)
```

### Writing a for Loop

```python
result = []

for value in df_ords_prods_merged['order_day_of_week'] :
    if value == 0:
        result.append('Busiest day')
    elif value == 4:
        result.append('Least Busy')
    else :
        result.append('Regularly Busy')
```

```python
result
```

```python
# creating a new column in df for busiest day
df_ords_prods_merged['busiest_day'] = result
```

```python
df_ords_prods_merged['busiest_day'].value_counts(dropna=False)
```

Classifying Assumption: > 2.5 Mil counts = Most orders, < 0.15 Mil counts = Fewest orders, the rest = average

```python
# creating loop for busiest hour labels
result_hour = []

for value in df_ords_prods_merged['order_hour_of_day']:
    if value in [10,11,14,15,13,12,16]:
        result_hour.append('Most orders')
    elif value in [3,4,2,5,1]:
        result_hour.append('Fewest orders')
    else:
        result_hour.append('Average orders')
```
